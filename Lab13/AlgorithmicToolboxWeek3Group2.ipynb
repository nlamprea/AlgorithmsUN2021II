#Money Change

The goal in this problem is to find the minimum numbers of coins needed to change the input value into coins with denominators 1, 5, and 10.

Approach: A common intuition would be to take coins with greater value first. This can reduce the total number of coins needed. Start from the largest possible denomination and keep adding denominations while the remaining value is greater than 0.

Constraints: $1 \leq n \leq 10^{3}$

**Sample:**

Imput: 28

Output: 6 (28 = 10 + 10 + 5 + 1 + 1 + 1)

```python
n = int(input())
count = 0
for i in [10, 5, 1]:
    if n>=i:
        q = n//i
        count += q
        n = n%i
        if n==0:
            print(count)
            quit()
```

#Maximum Value of the Loot (Fractional Knapsack)

The goal of this problem in to implement an algorithm for the knapsck problem.

Imput format: n, the number of compounds and W, the capacity of the backpack. The next n lines define the price and weight of each compound.

Output format: Maximum price of spices stuffed into the backpack.

Constraints: $1 \leq n \leq 10^{3}, W \leq 2*10^{6}, 0 \leq p_{i} \leq 2*10^{6}, 0 \leq w_{i} \leq 2*10^{6} for , 1 \leq i \leq n$


**Sample:**

Imput: 

2 50

60 20

100 50

120 30

Output: 180.0

```python
n, W = [int(i) for i in input().split()]
lst = []

if W == 0:
    print(0)
    quit()

for _ in range(n):
    v, w = [int(i) for i in input().split()]
    if v == 0:
        continue
    lst.append((v, w))

lst.sort(key = lambda x: x[0]/x[1], reverse = True)

total_value = 0

for v,w in lst:
    if W==0:
        print(total_value)
        quit()
    amt = min(w, W)
    total_value += amt*v/w
    W -= amt

print(total_value)
```

#Car Fueling

You are going to travel to another city that is located ùëë miles away from your home city. Your car can travel.
At most ùëö miles on a full tank and you start with a full tank. Along your way, there are gas stations at
Distances stop1, stop2, . . . , stopùëõ from your home city. What is the minimum number of refills needed?

Input Format. The first line contains an integer ùëë. The second line contains an integer ùëö. The third line.

Output Format: Assuming that the distance between the cities is ùëë miles, a car can travel at most ùëö miles on a full tank, and there are gas stations at distances stop1, stop2, . . . , stopùëõ along the way, output the Minimum number of refills needed. Assume that the car starts with a full tank. If it is not possible to. Reach the destination, output ‚àí1.

Constraints: $1 \leq d \leq 105.1 \leq m \leq 400.1 \leq n \leq 300.0 < stop1< stop2< ... < stopn< d$

**Sample:**

Imput: 

950

400

4

200 375 550 750

Output: 2

```python
def car_refill(dist,cap,n,stops):
    stops.insert(0,0)
    stops.append(dist)
    num_refill,curr_refill = 0,0
    while curr_refill <= n:
        last_refill = curr_refill
        while (curr_refill <= n and stops[curr_refill + 1] - stops[last_refill] <= cap):
            curr_refill += 1
        if curr_refill == num_refill :
            return -1
        if curr_refill <= n:
            num_refill +=1
    return num_refill
```

#Maximum Advertisement Revenue (Maximum Dot Product)

The dot product of two sequences a(1) , a(2) , . . . , a(n) and b(1) , b(2) , . . . , b(n) of the same length is equal to a(1) b(1) + a(2) b(2) + ¬∑ ¬∑ ¬∑ + a(n) b(n).

Task: The goal is, given two sequences a(1) , a(2) , . . . , a(n) and b(1) , b(2) , . . . , b(n) , find a permutation œÄ of the second sequence such that the dot product of a(1) , a(2) , . . . , a(n) and b(œÄ 1) , b(œÄ 2) , . . . , b(œÄ n) is minimum.

Input Format: The first line contains an integer n, the second one contains a sequence of integers a(1) , a(2) , . . . , a(n) , the third one contains a sequence of integers b(1) , b(2) , . . . , b(n) .

Output Format: Output the minimum possible dot product.

Constraints: 1 ‚â§ n ‚â§ 10^3 ; ‚àí10^5 ‚â§ a(i) , b(i) ‚â§ 10^5 for all 1 ‚â§ i ‚â§ n.

**Sample:**

Imput:

3

1 3 -5

-2 4 1

Output: 23 (3*4+1*1+(-5)*(-2))

```python
n = int(input())
a = [int(i) for i in input().split()]
b = [int(i) for i in input().split()]
a.sort()
b.sort()
ans = sum([a[i]*b[i] for i in range(n)])
print(ans)
```

#Collecting Signatures (Covering Segments by Points)

You are given a set of segments on a line and your goal is to mark as few points on a line as possible so that each segment contains at least one marked point.

Task. Given a set of n segments {[a(0) ,b(0) ],[a(1) ,b(1) ],‚Ä¶,[a(n)‚àí1 ,b(n)‚àí1 ]} with integer coordinates on a line, find the minimum number m of points such that each segment contains at least one point. That is, find a set of integers X of the minimum size such that for any segment [a(i) ,b(i) ] there is a point x ‚àà X such that a(i) ‚â§ x ‚â§ b(i) .

Input Format: The first line of the input contains the number n of segments. Each of the following n lines contains two integers a(i) and b(i) (separated by a space) defining the coordinates of endpoints of the i-th segment.


Output Format: Output the minimum number m of points on the first line and the integer coordinates of m points (separated by spaces) on the second line. You can output the points in any order. If there are many such sets of points, you can output any set. (It is not difficult to see that there always exist a set of points of the minimum size such that all the coordinates of the points are integers.)

Constraints. 1 ‚â§ n ‚â§ 100; 0 ‚â§ a(i) ‚â§ b(i) ‚â§ 10^9 for all 0 ‚â§ i < n.

**Sample:**

Imput:

3

1 3 

2 5

3 6

Output:

1

3

```python
n = int(input())
lst = []

for _ in range(n):
    a, b = [int(i) for i in input().split()]
    lst.append((a,b))

lst.sort(key = lambda x: x[1])

index = 0
coordinates = []

while index < n:
    curr = lst[index]
    while index < n-1 and curr[1]>=lst[index+1][0]:
        index += 1
    coordinates.append(curr[1])
    index += 1
print(len(coordinates))
print(' '.join([str(i) for i in coordinates]))
```

#Maximum Number of Prizes (Different Summands)

This is an example of a problem where a subproblem of the corresponding greedy algorithm is slightly distinct from the initial problem.

Task. The goal of this problem is to represent a given positive integer n as a sum of as many pairwise distinct positive integers as possible. That is, to find the maximum k such that n can be written as a(1) + a(2) + ¬∑¬∑¬∑ + a(k) where a(1) ,‚Ä¶,a(k) are positive integers and a(i) != a(j) for all 1 ‚â§ i < j ‚â§ k.

Input Format: The input consists of a single integer n.

Output Format: In the first line, output the maximum number k such that n can be represented as a sum of k pairwise distinct positive integers. In the second line, output k pairwise distinct positive integers that sum up to n

Constraints: $1 ‚â§ n ‚â§ 10^9$

**Sample:**

Imput:

8


Output:

3

1 2 5

```python
n = int(input())
if n == 1:
    print(1)
    print(1)
    quit()
W = n
prizes = []
for i in range(1, n):
    if W>2*i:
        prizes.append(i)
        W -= i
    else:
        prizes.append(W)
        break

print(len(prizes))
print(' '.join([str(i) for i in prizes]))
```

#Maximum Salary (Largest Number)

The goal in this problem is to tweak the above algorithm so that it works not only for single-digit numbers, but for arbitrary poritive integers.

Task: Compose the largest number out of a set of integers.

Input Format: The first line of the imput contains an integer n. The second line contains integers $a_{1}, a_{2},..., a_{n}$.

Output Format: Output the largest number that can be composed out of $a_{1}, a_{2},..., a_{n}$.

Constraints: $1 ‚â§ n ‚â§ 100; 1 ‚â§ a_{i} ‚â§ 10^3 $ for all $1‚â§i‚â§n$.

**Sample:**

Imput:

3

23 39 92

Output:

923923

```python
n = int(input())
lst = [int(i) for i in input().split()]


def IsGreaterOrEqual(digit, max_digit):
    return int(str(digit)+str(max_digit))>=int(str(max_digit)+str(digit))

def largestnumber(lst):
    answer = []

    while lst!=[]:
        max_digit = 0
        for digit in lst:
            if IsGreaterOrEqual(digit, max_digit):
                max_digit = digit
        answer.append(max_digit)
        lst.remove(max_digit)

    return answer

print(''.join([str(i) for i in largestnumber(lst)]))
```
