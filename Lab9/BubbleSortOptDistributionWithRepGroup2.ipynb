# Bubble sort optimized time complexity, and time complexity histograms for permutations of n elements  with repetions

## Best case - Increasingly sorted array 

$$t_b(n)=n-1=\Theta (n)$$

## Worst case - decreasingly sorted array ($$\frac{n^2-n}{2} \text{ comparisons} + \frac{n^2-n}{2} \ \text{swaps} $$)
$$t_w(n)=n^{2}-n=\Theta (n^{2})$$

## Average case over unifrom permutations of n different elements

$$t_a(n)\approx 0.72934022n^{2} - 1.2637766n + 0.85689742=\Theta (n^{2})$$

```python
from itertools import product
import numpy as np 

# Algoritmo Optimizado modificado de Geeks for Geeks: https://www.geeksforgeeks.org/bubble-sort/
def bubbleSortOptAllCounters(unsorted):
    intComps, intSwaps, intSteps, intPasses = 0, 0, 0, 0
    for i in range(len(unsorted) - 1):
        isSwapped = False
        intPasses += 1
        for j in range(0, n - i - 1):
            intComps += 1
            if unsorted[j] > unsorted[j + 1]:
                intSwaps += 1
                unsorted[j], unsorted[j + 1] = unsorted[j + 1], unsorted[j]
                isSwapped = True
        # Optimización: Salir si no hay swaps
        if not isSwapped:
            break
    intSteps = intSwaps + intComps

    return intComps, intSwaps, intSteps, intPasses
    
```

```python
def maxinversions(arr):
  maxinv = 0
  n = len(arr) 
  for i in range(n-1):
    inv = 0
    for j in range(i+1,n):
      if arr[j] < arr[i]:
        inv +=1
    if inv > maxinv:
      maxinv = inv
  return maxinv
```

```python
from itertools import product
import matplotlib.pyplot as plt
import numpy as np 
import math

n = 3
perm = product(range(1,n+1), repeat = n)

comps = []
swaps = []
steps = []
passes = [] 
maxinvs = []

for i in perm:
    i = list(i)
    maxinvs.append(maxinversions(i))
    intComps, intSwaps, intSteps, intPasses = bubbleSortOptAllCounters(i)
    comps.append(intComps)
    swaps.append(intSwaps)
    steps.append(intSteps)
    passes.append(intPasses)

print("n =",n)  
print("number of permutations n! =",math.factorial(n)) 
print()
   
b = np.array(range(min(maxinvs)-1,max(maxinvs)+2) ) +0.5
plt.hist(maxinvs,bins=b, color="red")
plt.title("Bubble Sort Maximum number of inversions")
plt.xlabel("Maximun number of inversions ")
plt.ylabel("Frequency")
plt.show()

print("Maximum number of inversions") 
print("Maximum number of inversions exper min =",min(maxinvs))
print("Maximum number of inversions formula min = 0")
print("Maximum number of inversions exper max =",max(maxinvs))
print("Maximum number of inversions formula max n-1 =",n-1)
print("Maximum number of inversions exper mean =",np.mean(maxinvs))
print("Maximum number of inversions formula mean  n- math.sqrt(math.pi*n/2)=",n- math.sqrt(math.pi*n/2))
print("Maximum number of inversions exper stdev =",np.std(maxinvs))
print()

b = np.array(range(min(comps)-1,max(comps)+2) ) +0.5
plt.hist(comps,bins=b, color="red")
plt.title("Bubble Sort Comparisons")
plt.xlabel("Comparisons ")
plt.ylabel("Frequency")
plt.show()

print("Comparisons") 
print("Comparisons formula (n-1)*n/2 =",(n-1)*n/2)
print("Comparisons exper min =",min(comps))
print("Comparisonsexper max =",max(comps))
print("Comparisons exper mean =",np.mean(comps))
print("Comparisons exper stdev =",np.std(comps))
print()

b = np.array(range(min(swaps)-1,max(swaps)+2) ) +0.5
plt.hist(swaps,bins=b,color="red")
plt.title("Bubble Sort Swaps")
plt.xlabel("Swaps ")
plt.ylabel("Frequency")
plt.show()

print("Swaps") 
print("Swaps exper min =",min(swaps))
print("Swaps min formula = 0")
print("Swaps exper max =",max(swaps))
print("Swaps max formula (n-1)*n/2 =",(n-1)*n/2)
print("Swaps exper mean =",np.mean(swaps))
print("Swaps mean formula (1/4*n**2)- (1/2*n)+(1/4) =",(1/4*n**2)- (1/2*n)+(1/4))
print("Swaps exper stdev =",np.std(swaps))
print()

b = np.array(range(min(steps)-1,max(steps)+2) ) +0.5
plt.hist(steps,bins=b,color="red")
plt.title("Bubble Sort Steps")
plt.xlabel("Steps ")
plt.ylabel("Frequency")
plt.show()

print("Steps") 
print("Steps exper min =",min(steps))
print("Steps min formula (n-1)*n/2 =",(n-1)*n/2)
print("Steps exper max =",max(steps))
print("Steps max formula (n-1)*n =",(n-1)*n)
print("Steps exper mean =",np.mean(steps))
print("Steps mean formula approximation (0.72934022*(n**2))+(-1.2637766*n) + 0.85689742 =",(0.72934022*(n**2))+(-1.2637766*n) + 0.85689742)
print("Steps exper stdev",np.std(steps))
print()

b = np.array(range(min(passes)-1,max(passes)+2) ) +0.5
plt.hist(passes,bins=b,color="red")
plt.title("Bubble Sort Passes")
plt.xlabel("Passes")
plt.ylabel("Frequency")
plt.show()
    
print("Passes") 
print("Passes formula (n-1) =",(n-1))
print("Passes exper min =",min(passes))
print("Passes exper max =",max(passes))
print("Passes exper mean =",np.mean(passes))
print("Passes exper stdev",np.std(passes))
```

```python
import numpy as np 
import pandas as pd
import scipy.linalg as la

n = np.array(list(range(9)))    # Number n of sorted elments 

t = []
for i in range(len(n)):
  t.append((0.73*n[i]**2)-(1.26*n[i])+0.86)

t = np.array(t) # Average comparisons + swaps for the corresponding n 

data = pd.DataFrame({'n': n, 't': t})
data
```
