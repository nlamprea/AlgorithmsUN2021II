# Bubble sort optimized time complexity, and time complexity histograms for permutations of n different elements 

## Best case - Increasingly sorted array 

$$t_b(n)=n-1=\Theta (n)$$

## Worst case - decreasingly sorted array ($$\frac{n^2-n}{2} \text{ comparisons} + \frac{n^2-n}{2} \ \text{swaps} $$)
$$t_w(n)=n^{2}-n=\Theta (n^{2})$$

## Average case over unifrom permutations of n different elements

$$ t_a(n)\approx \frac{57}{80}n^{2}-\frac{59}{80}n+\frac{2}{15}=\Theta (n^{2}) $$

```python
from itertools import product
import numpy as np 

# Algoritmo Optimizado modificado de Geeks for Geeks: https://www.geeksforgeeks.org/bubble-sort/
def bubbleSortOptAllCounters(unsorted):
    intComps, intSwaps, intSteps, intPasses = 0, 0, 0, 0
    for i in range(len(unsorted) - 1):
        isSwapped = False
        intPasses += 1
        for j in range(0, n - i - 1):
            intComps += 1
            if unsorted[j] > unsorted[j + 1]:
                intSwaps += 1
                unsorted[j], unsorted[j + 1] = unsorted[j + 1], unsorted[j]
                isSwapped = True
        if not isSwapped:
            break
    intSteps = intSwaps + intComps

    return intComps, intSwaps, intSteps, intPasses
    
```

```python
def maxinversions(arr):
  maxinv = 0
  n = len(arr) 
  for i in range(n-1):
    inv = 0
    for j in range(i+1,n):
      if arr[j] < arr[i]:
        inv +=1
    if inv > maxinv:
      maxinv = inv
  return maxinv
```

```python
from itertools import product
import numpy as np 

# https://www.geeksforgeeks.org/bubble-sort/

def bubbleSortOptSteps(unsorted):
    intComps = 0
    intSwaps = 0
    intSteps = 0
    intPasses = 0
    for i in range(len(unsorted) - 1):
        isSwapped = False
        intPasses += 1
        for j in range(0, n - i - 1):
            intComps += 1
            if unsorted[j] > unsorted[j + 1]:
                intSwaps += 1
                unsorted[j], unsorted[j + 1] = unsorted[j + 1], unsorted[j]
                isSwapped = True
        if not isSwapped:
            break
    intSteps = intSwaps + intComps

    return intSteps
    
```
